```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

#Basic Graph

```python
x = np.arange(0,20,0.2)
y = np.sin(x)

plt.figure(figsize=(11,4),dpi=125)

plt.plot(x,y,color = 'b',label = 'sin(x)')
plt.plot(x,np.log(x),'g',label='log(x)')
plt.title('Sine Graph')
plt.xlabel('X')
plt.ylabel('sin(X)')
plt.legend()
plt.savefig('graph.png',dpi=300)
plt.show()
```

#Bar Graphs

```python
plt.figure(figsize=(10,4),dpi=140)

x = ['W1','W2','W3','W4','W5']
y1 = [21.2,34.4,23.1,27.8,32.2]
y2 = [21.9,23.5,27.4,26.7,28.6]
y3 = [31.2,29.8,25.6,33.4,36.7]

x_pos = np.arange(len(x))
bar_width = 0.2

plt.bar(x_pos,y1,bar_width,label='Price_Stock_1')
plt.bar(x_pos-bar_width,y2,bar_width,label='Price_Stock_2')
plt.bar(x_pos+bar_width,y3,bar_width,label='Price_Stock_3')
plt.legend()
plt.title('Variation of stock prices')
plt.xlabel('Weeks')
plt.ylabel('Prices')
plt.xticks(x_pos,x)
plt.show()
```

```python
plt.figure(figsize=(11,4),dpi=120)

x = np.arange(-10,10,0.2)
y = np.sin(x)/x

plt.stem(x,y,label='sinc(x)',basefmt='g')
plt.xlabel('X')
plt.ylabel('sinc(X)')
plt.title('sinc(x)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(11,4),dpi=120)
x = np.arange(10)
l = [21,12,32,11,23,54,21,50,32,13]
m = [44,33,23,54,34,38,29,54,23,64]
n = [34,56,81,23,54,60,32,17,45,23]

labels = ['Area 1','Area 2','Area 3']
plt.stackplot(x,l,m,n,labels=labels)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend()
plt.show()
```
